Carregar o pacote pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Carregar o dataset

```python
df = pd.read_csv('dataset/spotify_dataclean.csv')
df['Track'] = df['artist_name'] + '\n' + df['track_name']
```

Filtrando pra ter só musica boa

```python
df = df[df['artist_genres'].str.contains('rock', case=False, na=False)]
```

1. Insight de hits com maior duração

```python
head_max_duration = df[['Track', 'track_duration_min']].sort_values(by='track_duration_min', ascending=False).head().reset_index(drop=True)
```

2. Insight de hits com maior popularidade

```python
head_max_popularity = df[['Track', 'track_popularity', 'explicit']].sort_values(by='track_popularity', ascending=False).head().reset_index(drop=True)
```

**Seaborn Configurations**

```python
sns.set_palette("plasma")
sns.set_context('notebook')
```

1. Grafico de popularidade 

```python
graf_max_popularity = sns.lineplot(data=head_max_popularity, x='Track', y='track_popularity')
graf_max_popularity.tick_params(axis='x', labelsize=8)
#plt.savefig('insights/graf_max_popularity.png', dpi=300, bbox_inches='tight')
```

```python
graf_max_duration = sns.barplot(data=head_max_duration, x='track_duration_min', y='Track')
graf_max_popularity.tick_params(axis='x', labelsize=8 )
#plt.savefig('insights/graf_max_popularity.png', dpi=300, bbox_inches='tight')
```
